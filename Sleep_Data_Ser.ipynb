```
import pandas as pd
import matplotlib.pyplot as plt

file_path = "sleep.csv"
sleep = pd.read_csv(file_path)

stress = sleep['Stress Level']
physical_activity = sleep['Physical Activity Level']

plt.figure(figsize=(8, 6))
plt.scatter(stress, physical_activity)
plt.title("Scatter Plot of Stress vs. Physical Activity")
plt.xlabel("Stress")
plt.ylabel("Physical Activity")
plt.show()

correlation = sleep['Stress Level'].corr(sleep['Physical Activity Level'])
print(f"Correlation between Stress and Physical Activity: {correlation}")
```
