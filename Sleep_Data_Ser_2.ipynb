```
import pandas as pd
import matplotlib.pyplot as plt

file_path = "sleep.csv"
sleep = pd.read_csv(file_path)
count_obese_apnea = len(sleep[(sleep['BMI Category'] == 'Obese') & (sleep['Sleep Disorder'] == 'Sleep Apnea')])
print(f'The number of people who have both sleep apnea and are obese is: {count_obese_apnea}')


sleep = pd.DataFrame({
    'BMI Category': ['Obese', 'Normal', 'Overweight', 'Obese', 'Normal', 'Overweight'],
    'Sleep Disorder': ['None', 'None', 'None', 'Sleep Apnea', 'Sleep Apnea', 'Sleep Apnea'],
     'Count': [1, 2, 3, 4, 5, 6]
})

grouped_data = sleep.groupby(['BMI Category', 'Sleep Disorder']).size().unstack().fillna(0)

ax = grouped_data.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title("BMI Category vs. Sleep Disorder")
plt.xlabel("BMI Category")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title="Sleep Disorder")

plt.show()








```
